# Import

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
import matplotlib.animation as animation
from IPython.display import HTML, display
from matplotlib.animation import FuncAnimation
from scipy import signal
```

# Function 

```python

def f(t):
  return (5*np.sin(t)+2*np.sin(2*t)+6*np.sin(3*t))
```

# Play 1

```python

T=np.linspace(-10,10,1000)
F= f(T)

T.shape, F.shape
```

```python
plt.figure(figsize=(20,7.5))
plt.axhline(y=0, linestyle='-')
plt.plot(T,5*np.sin(T),"green",label="component 1")
plt.plot(T,2*np.sin(2*T),"red",label="component 2")
plt.plot(T,6*np.sin(3*T),"blue",label="component 3")
plt.ylim([-7, 7])
plt.title("The sample function along with its different frequency components")
plt.legend()

plt.figure(figsize=(20,7.5))
plt.axhline(y=0, linestyle='-')
plt.plot(T,F,"black",label="f(t)")
plt.ylim([-14, 14])
plt.show()
```

```python
plt.figure(figsize=(25,10))

plt.subplot(221)
plt.plot(T,F,"k")
plt.title("f(t)")

plt.subplot(222)
plt.plot(T,5*np.sin(T),"indigo",label="component 1")
plt.title("component 1")

plt.subplot(223)
plt.plot(T,2*np.sin(2*T),"red",label="component 2")
plt.title("component 2")

plt.subplot(224)
plt.plot(T,6*np.sin(3*T),"blue")
plt.title("component 3")

plt.show()
```

```python
plt.figure(figsize=(25,10))

t = np.linspace(0, 100, 5000, endpoint=False)
square = signal.square(np.pi*t)

plt.plot(t, square,"k")
plt.ylim(-2, 2)
plt.xlim(0,10)
plt.axhline(y=0)
plt.title("Rectangular wave")
plt.show()
```

```python
t1=np.linspace(0,100,50_000)

f=[]
for i in range(8):
  f.append([])
  for t in t1:
    f[i].append(4*np.sin((2*i+1)*np.pi*t)/((2*i+1)*np.pi))

plt.figure(figsize=(25,12))
plt.subplot(211)
for i in range(8):
  plt.plot(t1,f[i])
  plt.axhline(y=0)
  plt.xlim(0,3)
plt.title("Indivual componets",fontsize=15)

fn=f[0]
for i in range(1,8):
  fn=np.add(fn,f[i])

plt.subplot(212)
plt.plot(t1,fn,"k")
plt.axhline(y=0)
plt.xlim(0,3)
plt.title("f(t) as a summation of first 8 harmonics",fontsize=15)
plt.show()
```

```python
t1=np.linspace(0,100,50_000)

def rectapp(n):
  for i in range(n):
    f=[]
    for i in range(n):
      f.append([])
      for t in t1:
        f[i].append(4*np.sin((2*i+1)*np.pi*t)/((2*i+1)*np.pi))

  fn=f[0]
  # plt.figure(figsize=(15,9))
  plt.plot(t1,fn,"k")
  plt.axhline(y=0)
  plt.xlim(0,3)
  plt.title("Approximation with 1 frequency component")

  for i in range(1,n):
    fn=np.add(fn,f[i])
    plt.subplot(n,1,i+1)
    plt.figure(figsize=(7, 20))
    for j in range(i+1):
      plt.plot(t1,f[j])
    plt.plot(t1,fn,"k")
    plt.axhline(y=0)
    plt.xlim(0,3)
    plt.title("Approximation with "+str(i+1)+" frequency components")

rectapp(10)
```
